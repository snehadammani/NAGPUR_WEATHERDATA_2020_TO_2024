```python
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# 1. Load prediction CSVs
# -----------------------------
rf_df = pd.read_csv("randomforest_predictions.csv")
xgb_df = pd.read_csv("xgboost_predictions.csv")

# Convert date column
rf_df['date'] = pd.to_datetime(rf_df['date'])
xgb_df['date'] = pd.to_datetime(xgb_df['date'])

# -----------------------------
# 2. Plot Max Temperature Comparison
# -----------------------------
plt.figure(figsize=(10, 5))

plt.plot(rf_df['date'], rf_df['Pred_MaxTemp'], marker='o', label='RandomForest')
plt.plot(xgb_df['date'], xgb_df['Pred_MaxTemp'], marker='*', label='XGBoost')

plt.xlabel("Date")
plt.ylabel("Predicted Max Temperature (C)")
plt.title("RandomForest vs XGBoost  Max Temperature Prediction")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# 1. Load prediction CSV files
# -----------------------------
rf = pd.read_csv("randomforest_predictions.csv")
xgb = pd.read_csv("xgboost_predictions.csv")
lstm = pd.read_csv("lstm_predictions.csv")

# Convert date column to datetime
rf['date'] = pd.to_datetime(rf['date'])
xgb['date'] = pd.to_datetime(xgb['date'])
lstm['date'] = pd.to_datetime(lstm['date'])

# -----------------------------
# 2. Plot Max Temperature Comparison
# -----------------------------
plt.figure(figsize=(11, 6))

plt.plot(
    rf['date'], rf['Pred_MaxTemp'],
    marker='o', linestyle='-',
    label='RandomForest',
)

plt.plot(
    xgb['date'], xgb['Pred_MaxTemp'],
    marker='s', linestyle='--',
    label='XGBoost',
)

plt.plot(
    lstm['date'], lstm['Pred_MaxTemp'],
    marker='^', linestyle='-.',
    label='LSTM',
)

plt.xlabel("Date")
plt.ylabel("Predicted Max Temperature (C)")
plt.title("RF vs XGBoost vs LSTM  Max Temperature Prediction Comparison")

plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# 1. Load LSTM predictions
# -----------------------------
df = pd.read_csv("2026_lstm_predictions.csv")
df['date'] = pd.to_datetime(df['date'])

# -----------------------------
# 2. Plot Max & Min Temperature
# -----------------------------
plt.figure(figsize=(11, 6))

plt.plot(
    df['date'],
    df['Pred_MaxTemp'],
    marker='o',
    linestyle='-',
    color='red',
    label='Predicted Max Temperature'
)

plt.plot(
    df['date'],
    df['Pred_MinTemp'],
    marker='s',
    linestyle='--',
    color='blue',
    label='Predicted Min Temperature'
)

plt.xlabel("Date")
plt.ylabel("Temperature (C)")
plt.title("LSTM Weather Prediction (1 Jan 2026  7 Jan 2026)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
