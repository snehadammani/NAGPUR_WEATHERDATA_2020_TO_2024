```python
# Datetime column (already correct, but safe to re-run)
df['datetime'] = pd.to_datetime(df['datetime'], format='%d-%m-%Y')

# Sunrise & Sunset: handle mixed / wrong formats safely
df['sunrise'] = pd.to_datetime(df['sunrise'], errors='coerce').dt.time
df['sunset'] = pd.to_datetime(df['sunset'], errors='coerce').dt.time

# Check result
print("🔹 Updated Dataset Info after fixing time columns:")
df.info()

print("\n🔹 Sunrise & Sunset sample:")
display(df[['sunrise', 'sunset']].head(10))
```

```python
df.dtypes
```

```python
# Check missing values count
missing_count = df.isnull().sum()

# Check missing values percentage
missing_percent = (missing_count / len(df)) * 100

# Combine into a single DataFrame
missing_df = pd.DataFrame({
    'Missing_Count': missing_count,
    'Missing_Percentage': missing_percent
})

# Show only columns with missing values
missing_df = missing_df[missing_df['Missing_Count'] > 0]

print("🔹 Missing Value Analysis:")
display(missing_df)
```

```python
# Fill missing values in preciptype with 'none'
df['preciptype'] = df['preciptype'].fillna('none')

# Verify missing values again
print("🔹 Missing values after filling preciptype:")
print(df.isnull().sum()[['preciptype']])

# Check unique values in preciptype
print("\n🔹 Unique values in preciptype:")
print(df['preciptype'].value_counts())
```

```python
# Drop sunrise column
df = df.drop(columns=['sunrise'])

# Verify removal
print("🔹 Columns after dropping sunrise:")
print(df.columns)

# Final missing value check
print("\n🔹 Missing values after dropping sunrise:")
print(df.isnull().sum().sum())
```

```python
# Identify object (categorical) columns
object_cols = df.select_dtypes(include='object').columns

print("🔹 Object Columns:")
print(object_cols)

# Apply value_counts on each object column
for col in object_cols:
    print("\n" + "="*50)
    print(f"🔹 Value counts for column: {col}")
    print(df[col].value_counts())
```

```python
# 1️⃣ Statistical summary
print("🔹 Statistical Summary:")
display(df.describe())
```

```python

# 2️⃣ Distribution plots for key weather variables
key_cols = ['temp', 'humidity', 'pressure', 'windspeed', 'precip']

plt.figure(figsize=(15, 10))

for i, col in enumerate(key_cols, 1):
    plt.subplot(2, 3, i)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')

plt.tight_layout()
plt.show()
```

```python
# 1️⃣ Count of Rain vs No-Rain days
plt.figure(figsize=(5,4))
sns.countplot(x='preciptype', data=df)
plt.title("Rain vs No-Rain Days Count")
plt.xlabel("Precipitation Type")
plt.ylabel("Number of Days")
plt.show()
```

```python
# 2️⃣ Temperature comparison
plt.figure(figsize=(6,4))
sns.boxplot(x='preciptype', y='temp', data=df)
plt.title("Temperature on Rain vs No-Rain Days")
plt.show()
```

```python
# 3️⃣ Humidity comparison
plt.figure(figsize=(6,4))
sns.boxplot(x='preciptype', y='humidity', data=df)
plt.title("Humidity on Rain vs No-Rain Days")
plt.show()
```

```python
# 4️⃣ Pressure comparison
plt.figure(figsize=(6,4))
sns.boxplot(x='preciptype', y='pressure', data=df)
plt.title("Pressure on Rain vs No-Rain Days")
plt.show()
```

```python
# Create month and year columns from datetime
df['month'] = df['datetime'].dt.month
df['year'] = df['datetime'].dt.year

# Month-wise averages
monthly_avg = df.groupby('month')[['temp', 'humidity', 'precip']].mean().reset_index()
```

```python
# Plot monthly temperature trend
plt.figure(figsize=(6,4))
sns.lineplot(x='month', y='temp', data=monthly_avg, marker='o')
plt.title('Average Monthly Temperature')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.show()
```

```python
# Plot monthly humidity trend
plt.figure(figsize=(6,4))
sns.lineplot(x='month', y='humidity', data=monthly_avg, marker='o')
plt.title('Average Monthly Humidity')
plt.xlabel('Month')
plt.ylabel('Humidity (%)')
plt.show()
```

```python
# Plot monthly precipitation trend
plt.figure(figsize=(6,4))
sns.lineplot(x='month', y='precip', data=monthly_avg, marker='o')
plt.title('Average Monthly Precipitation')
plt.xlabel('Month')
plt.ylabel('Precipitation (mm)')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
corr_matrix = numeric_df.corr()

plt.figure(figsize=(18, 14))
sns.heatmap(
    corr_matrix,
    cmap='coolwarm',
    annot=True,        # 👈 IMPORTANT (show numbers)
    fmt=".2f",         # 2 decimal places
    linewidths=0.5
)

plt.title("Correlation Heatmap with Values")
plt.show()
```

```python
# Key numeric columns where outliers usually exist
outlier_cols = ['temp', 'humidity', 'windspeed', 'precip', 'pressure']

plt.figure(figsize=(15, 8))

for i, col in enumerate(outlier_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}')

plt.tight_layout()
plt.show()
```

```python
import numpy as np

# 1️⃣ Cap windspeed at 99th percentile
wind_cap = df['windspeed'].quantile(0.99)
df['windspeed_capped'] = np.where(
    df['windspeed'] > wind_cap,
    wind_cap,
    df['windspeed']
)

print("🔹 Windspeed 99th percentile cap value:", wind_cap)
print("🔹 Max windspeed after capping:", df['windspeed_capped'].max())


# 2️⃣ Log transform precipitation (to handle skewness)
df['precip_log'] = np.log1p(df['precip'])

print("\n🔹 Precipitation before vs after log transform:")
display(df[['precip', 'precip_log']].describe())
```
